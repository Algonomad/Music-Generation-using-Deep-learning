```python
import collections
import pathlib
import numpy as np
import pandas as pd
import pretty_midi
import tensorflow as tf
from matplotlib import pyplot as plt
from typing import Optional

# Set random seed for reproducibility
SEED = 42
tf.random.set_seed(SEED)
np.random.seed(SEED)
SAMPLE_RATE = 16000

# Dataset path
DATA_DIR = pathlib.Path('data/maestro-v2.0.0')
if not DATA_DIR.exists():
    tf.keras.utils.get_file(
        'maestro-v2.0.0-midi.zip',
        origin='https://storage.googleapis.com/magentadata/datasets/maestro/v2.0.0/maestro-v2.0.0-midi.zip',
        extract=True,
        cache_dir='.',
        cache_subdir='data',
    )

def parse_midi_file(midi_path: str) -> pd.DataFrame:
    """Parse MIDI file and return note attributes as a DataFrame."""
    midi_data = pretty_midi.PrettyMIDI(midi_path)
    notes = collections.defaultdict(list)
    instrument = midi_data.instruments[0]
    sorted_notes = sorted(instrument.notes, key=lambda note: note.start)
    previous_start = sorted_notes[0].start

    for note in sorted_notes:
        notes['pitch'].append(note.pitch)
        notes['start'].append(note.start)
        notes['end'].append(note.end)
        notes['step'].append(note.start - previous_start)
        notes['duration'].append(note.end - note.start)
        previous_start = note.start

    return pd.DataFrame(notes)

def plot_notes(notes: pd.DataFrame, count: Optional[int] = None):
    """Plot piano roll representation of notes."""
    plt.figure(figsize=(20, 4))
    count = count or len(notes['pitch'])
    plt.plot(
        [notes['start'], notes['end']],
        [notes['pitch'], notes['pitch']],
        color="b", marker=".")
    plt.xlabel('Time (s)')
    plt.ylabel('Pitch')
    plt.title(f"{'First ' + str(count) if count else 'All'} notes")
    plt.show()

def create_midi_from_notes(notes: pd.DataFrame, output_path: str, instrument_name: str) -> pretty_midi.PrettyMIDI:
    """Generate MIDI file from note data."""
    midi = pretty_midi.PrettyMIDI()
    instrument = pretty_midi.Instrument(
        program=pretty_midi.instrument_name_to_program(instrument_name)
    )
    prev_start = 0
    for _, note in notes.iterrows():
        start = float(prev_start + note['step'])
        end = float(start + note['duration'])
        midi_note = pretty_midi.Note(
            velocity=100,
            pitch=int(note['pitch']),
            start=start,
            end=end,
        )
        instrument.notes.append(midi_note)
        prev_start = start

    midi.instruments.append(instrument)
    midi.write(output_path)
    return midi

def preprocess_notes(filenames, num_files: int, seq_len: int) -> tf.data.Dataset:
    """Convert notes into a sequence dataset."""
    notes_data = []
    for file in filenames[:num_files]:
        notes_data.append(parse_midi_file(str(file)))  # Convert PosixPath to string
    notes_df = pd.concat(notes_data)

    notes_array = np.stack(
        [notes_df['pitch'], notes_df['step'], notes_df['duration']], axis=1
    )
    notes_ds = tf.data.Dataset.from_tensor_slices(notes_array)

    def window_fn(dataset: tf.data.Dataset, window_size: int):
        windows = dataset.window(window_size + 1, shift=1, drop_remainder=True)
        sequences = windows.flat_map(lambda x: x.batch(window_size + 1))

        def split_fn(seq):
            inputs = seq[:-1]
            targets = {key: seq[-1][i] for i, key in enumerate(['pitch', 'step', 'duration'])}
            return inputs, targets

        return sequences.map(split_fn)

    return window_fn(notes_ds, seq_len)


# Load data
file_paths = list(DATA_DIR.glob('**/*.mid')) + list(DATA_DIR.glob('**/*.midi'))
if not file_paths:
    raise FileNotFoundError(f"No MIDI files found in {DATA_DIR}. Check the dataset path.")

sample_file = file_paths[0]
sample_notes = parse_midi_file(str(sample_file))

# Plot notes
plot_notes(sample_notes, count=100)

# Prepare dataset
SEQUENCE_LENGTH = 25
BATCH_SIZE = 64
NUM_FILES = 5

dataset = preprocess_notes(file_paths, NUM_FILES, SEQUENCE_LENGTH)
dataset = dataset.shuffle(10000).batch(BATCH_SIZE).cache().prefetch(tf.data.AUTOTUNE)

# Define model
input_layer = tf.keras.Input(shape=(SEQUENCE_LENGTH, 3))
lstm = tf.keras.layers.LSTM(128)(input_layer)
outputs = {
    'pitch': tf.keras.layers.Dense(128, activation='softmax', name='pitch')(lstm),
    'step': tf.keras.layers.Dense(1, activation='relu', name='step')(lstm),
    'duration': tf.keras.layers.Dense(1, activation='relu', name='duration')(lstm),
}

model = tf.keras.Model(inputs=input_layer, outputs=outputs)
losses = {
    'pitch': tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    'step': 'mse',
    'duration': 'mse',
}
model.compile(optimizer='adam', loss=losses, loss_weights={'pitch': 0.05, 'step': 1.0, 'duration': 1.0})
model.summary()

# Train model
callbacks = [
    tf.keras.callbacks.ModelCheckpoint(
        filepath='./checkpoints/model_epoch_{epoch}.weights.h5',  # Add .weights.h5 extension
        save_weights_only=True
    ),
    tf.keras.callbacks.EarlyStopping(
        monitor='loss', patience=5, restore_best_weights=True
    ),
]

history = model.fit(dataset, epochs=50, callbacks=callbacks)

# Plot training history
plt.plot(history.epoch, history.history['loss'], label='Loss')
plt.legend()
plt.show()

# Generate notes
def generate_notes(model, seed_notes: np.ndarray, num_predictions: int, temperature: float = 1.0):
    generated = []
    input_notes = seed_notes / np.array([128, 1, 1])
    prev_start = 0

    for _ in range(num_predictions):
        input_tensor = tf.expand_dims(input_notes, 0)
        predictions = model.predict(input_tensor)
        pitch = tf.random.categorical(predictions['pitch'] / temperature, num_samples=1)
        step = predictions['step'][0, 0]
        duration = predictions['duration'][0, 0]

        pitch = tf.squeeze(pitch).numpy()
        step = max(0, step)
        duration = max(0, duration)

        start = prev_start + step
        end = start + duration
        generated.append((pitch, step, duration, start, end))

        input_notes = np.append(input_notes[1:], [[pitch, step, duration]], axis=0)
        prev_start = start

    return pd.DataFrame(generated, columns=['pitch', 'step', 'duration', 'start', 'end'])

# Generate and save MIDI

seed_sequence = sample_notes[['pitch', 'step', 'duration']].values[:SEQUENCE_LENGTH]
new_notes = generate_notes(model, seed_sequence, num_predictions=100, temperature=0.8)
new_midi = create_midi_from_notes(new_notes, 'generated_output.midi', instrument_name='Acoustic Grand Piano')
```

```python
from IPython.display import Audio
import pandas as pd

def play_generated_notes(generated_notes: pd.DataFrame, instrument_name: str, duration_sec: int = 30):
    """Generate and play audio from predicted notes."""
    # Create a PrettyMIDI object from the generated notes
    midi = create_midi_from_notes(generated_notes, 'temp_output.midi', instrument_name)

    # Synthesize the audio
    waveform = midi.fluidsynth(fs=SAMPLE_RATE)

    # Trim the audio to the specified duration
    waveform_short = waveform[:duration_sec * SAMPLE_RATE]

    # Return an IPython Audio object for playback
    return Audio(waveform_short, rate=SAMPLE_RATE)

# Play the generated notes for 30 seconds
play_generated_notes(new_notes, instrument_name='Acoustic Grand Piano', duration_sec=30)
```
